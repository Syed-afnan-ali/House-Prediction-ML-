```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
```

```python
df=pd.read_csv("housing.csv")
```

```python
df
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(['median_house_value'], axis=1)
Y=df["median_house_value"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
train_data=X_train.join(Y_train)
```

```python
train_data
```

```python
correlation_matrix = train_data.corr(numeric_only=True)
```

```python
correlation_matrix
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(numeric_only=True),annot=True,cmap="YlGnBu")
```

```python
train_data['total_rooms']=np.log(train_data['total_rooms']+1)
train_data['total_bedrooms']=np.log(train_data['total_bedrooms']+1)
train_data['population']=np.log(train_data['population']+1)
train_data['households']=np.log(train_data['households']+1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data.ocean_proximity.value_counts()
```

```python
train_data = pd.concat([train_data.drop('ocean_proximity', axis=1), pd.get_dummies(train_data['ocean_proximity'], prefix='ocean')], axis=1)
```

```python
train_data
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(numeric_only=True),annot=True,cmap="YlGnBu")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="latitude", y="longitude", data=train_data, hue="median_house_value", palette="coolwarm")
```

```python
train_data['bedrooms_ratio']=train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_rooms']=train_data['total_rooms']/train_data['households']
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(numeric_only=True),annot=True,cmap="YlGnBu")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler


scaler=StandardScaler()

X_train = train_data.drop(['median_house_value'], axis=1)
Y_train = train_data['median_house_value']

X_train_s = scaler.fit_transform(X_train)
reg = LinearRegression()
reg.fit(X_train_s, Y_train)
```

```python
test_data=X_test.join(Y_test)
test_data['total_rooms']=np.log(test_data['total_rooms']+1)
test_data['total_bedrooms']=np.log(test_data['total_bedrooms']+1)
test_data['population']=np.log(test_data['population']+1)
test_data['households']=np.log(test_data['households']+1)
test_data = pd.concat([test_data.drop('ocean_proximity', axis=1), pd.get_dummies(test_data['ocean_proximity'], prefix='ocean')], axis=1)
test_data['bedrooms_ratio']=test_data['total_bedrooms']/test_data['total_rooms']
test_data['household_rooms']=test_data['total_rooms']/test_data['households']

X_test = test_data.drop(['median_house_value'], axis=1)
Y_test = test_data['median_house_value']
```

```python
X_test_s=scaler.transform(X_test)
```

```python
reg.score(X_test_s,Y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
forest=RandomForestRegressor()
forest.fit(X_train_s,Y_train)
```

```python
forest.score(X_test_s,Y_test)
```

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor(random_state=42) # Add random_state for reproducibility
param_grid = {
    "n_estimators": [100, 200],
    "min_samples_split": [2, 4],
    "max_depth": [None, 4, 8]
}
grid_search = GridSearchCV(forest, param_grid, cv=5, scoring="neg_mean_squared_error", return_train_score=True, n_jobs=-1, verbose=2)

grid_search.fit(X_train_s, Y_train)
```

```python
grid_search.best_estimator_.score(X_test_s,Y_test)
```
